```python
survived_count=df[df["Survived"]==1]["Sex"].value_counts().sort_index()
not_survived_count=df[df["Survived"]==0]["Sex"].value_counts().sort_index()
```

```python
plt.style.use('dark_background')
plt.plot(survived_count.index,survived_count.values,marker="o",color="green",linestyle='-')
plt.plot(not_survived_count.index,survived_count.values,marker="p",color="red",linestyle='-')
plt.ylabel("count ",weight='bold')
plt.xlabel('sex',weight='bold')
plt.title("survival count by sex ",weight='bold')
plt.legend()
```

```python
women=df[(df['Sex']=='female')&(df['Survived']==1)]

men=df[(df['Sex']=='male')&(df['Survived']==1)]
```
